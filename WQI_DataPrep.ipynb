```python
import pandas as pd
prod_pd = pd.read_csv('WTP_Monthly_Prod_2021_2022.csv')
prod_pd = prod_pd.melt(id_vars=['wtp'], var_name='Month', value_name='Production')
prod_pd['Month'] = pd.to_datetime(prod_pd['Month'], format='%d/%m/%Y').dt.strftime('%m-%Y')
prod_pd[prod_pd['wtp'] == 'Ampang Intake']
```

```python
expense_pd = pd.read_csv('WTP Monthly Detail Chemical Expenses Report 2022_December Latest (003).csv')
expense_pd['Month'] = pd.to_datetime(expense_pd['Date'], format='%d/%m/%Y').dt.strftime('%m-%Y')
expense_sum = expense_pd.groupby(['wtp', 'Month'])['TotalRMAmount'].sum().reset_index()
```

```python
expense_pd
```

```python
expense_sum
```

```python
unit_cost_pd = pd.merge(expense_sum,prod_pd, on=['wtp','Month'], how = 'outer')
unit_cost_pd['Production'] = unit_cost_pd['Production'].astype(float)
unit_cost_pd['TotalRMAmount'] = unit_cost_pd['TotalRMAmount'].astype(float)
unit_cost_pd['Unit_cost'] = (unit_cost_pd['TotalRMAmount']/unit_cost_pd['Production'])/10
unit_cost_pd.to_csv('unit_cost_V2_2022.csv')
```
