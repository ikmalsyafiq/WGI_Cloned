```python
import pandas as pd
import numpy as np
```

```python
#df = pd.read_csv('2020 Data Raw Water Quality.csv')
df = pd.read_csv('2022 Raw Water Parameter.csv')
```

```python
columns = ['WTP','TestDate','RawWaterTotalViolation','wqc_raw_total_avg']
```

```python
test = df[columns].copy()
kk = pd.pivot_table(test, index=['WTP','TestDate'] , columns=['RawWaterTotalViolation'], values=['wqc_raw_total_avg'])
kk.columns = kk.columns.droplevel(0)
k2 = kk.reset_index().rename_axis(None, axis=1)
k2.describe()
```

```python
cdict = {
    'WTP':'wtp',
    'TestDate':'Date',
    'Aluminium (Al)':'Aluminium_avg',
    'Ammonia (NH4)':'Ammonia_avg',
    'Color':'Color_avg',
    'Combined Chlorine':'Chlorine_avg',
    'E.coli':'E.coli_avg',
    'Ferum/Iron (Fe)':'Iron_avg',
    'Fluoride':'Fluoride_avg',
    'Free Residual Chlorine':'FRChlorine_avg',
    'Kadar Aliran':'flow_rate_avg',
    'Manganese (Mn)':'Manganese_avg',
    'Odor':'Odor_avg',
    'Taste':'Taste_avg',
    'Total coliform':'Total_coliform_avg',
    'Turbidity':'Turbidity_avg',
    'Weather':'Weather_avg',
    'pH':'pH_avg'
}
```

```python
k3 = k2.rename(columns=cdict)
```
